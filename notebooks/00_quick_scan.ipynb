```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Load dataset ---
df = pd.read_csv("../data/train_test_network.csv")

print("Shape:", df.shape)
print("\nDtypes:\n", df.dtypes)

# --- Preview rows ---
display(df.head(20))  # first 20 rows are enough to check schema
display(df.sample(20, random_state=42))  # random rows (better check for anomalies)

# --- Summary stats ---
display(df.describe(include='all').T)

# --- Missing values ---
null_counts = df.isna().sum().sort_values(ascending=False)
print("\nTop 30 columns by missing values:\n", null_counts.head(30))
sns.heatmap(df.isna(), cbar=False)
plt.title("Missingness Heatmap")
plt.show()

# --- Timestamp check ---
for t in ['timestamp', 'time', 'ts', 'Date']:
    if t in df.columns:
        print(f"{t} range:", df[t].min(), "→", df[t].max())

# --- Label column detection ---
LABEL_COL = None
for cand in ['label','Label','attack','class']:
    if cand in df.columns:
        LABEL_COL = cand
        print("Label column found:", cand)
        print(df[cand].value_counts().head(20))
        break

# --- Normalize label column (if found) ---
if LABEL_COL:
    df[LABEL_COL] = df[LABEL_COL].astype(str).str.strip()
    df[LABEL_COL] = df[LABEL_COL].replace({'BENIGN':'Normal','benign':'Normal'})
    print("\nCleaned label distribution:\n", df[LABEL_COL].value_counts())
```
