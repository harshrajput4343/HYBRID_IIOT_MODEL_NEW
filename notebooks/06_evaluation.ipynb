```python
import os
import matplotlib.pyplot as plt
from sklearn.metrics import (
    confusion_matrix, ConfusionMatrixDisplay,
    classification_report, roc_curve, auc
)
import pandas as pd
import numpy as np

def evaluate_and_save(y_true, y_pred, name="hybrid_fusion"):
    """
    Evaluates the model predictions and saves all evaluation outputs
    into results/eval/ directory.
    """
    # Ensure eval folder exists
    eval_dir = "results/eval"
    os.makedirs(eval_dir, exist_ok=True)

    # --- Confusion Matrix ---
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap="Blues", xticks_rotation=45)
    plt.title(f"Confusion Matrix - {name}")
    plt.tight_layout()
    plt.savefig(f"{eval_dir}/{name}_confusion_matrix.png", dpi=300)
    plt.close()
    print(f"[INFO] Confusion matrix saved at {eval_dir}/{name}_confusion_matrix.png")

    # --- Classification Report ---
    report = classification_report(y_true, y_pred, output_dict=True)
    report_df = pd.DataFrame(report).transpose()
    report_df.to_csv(f"{eval_dir}/{name}_classification_report.csv")
    
    # Also save as TXT (pretty formatted)
    with open(f"{eval_dir}/{name}_classification_report.txt", "w") as f:
        f.write(classification_report(y_true, y_pred))
    
    print(f"[INFO] Classification report saved at {eval_dir}/{name}_classification_report.csv and .txt")

    # --- ROC Curve ---
    # NOTE: ROC curve is primarily for binary classification.
    # For multi-class, we need a one-vs-rest strategy or macro-average.
    try:
        from sklearn.preprocessing import label_binarize
        classes = sorted(list(set(y_true)))
        y_true_bin = label_binarize(y_true, classes=classes)
        y_pred_bin = label_binarize(y_pred, classes=classes)

        fpr, tpr, _ = roc_curve(y_true_bin.ravel(), y_pred_bin.ravel())
        roc_auc = auc(fpr, tpr)

        plt.figure()
        plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
        plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.title(f"ROC Curve - {name}")
        plt.legend()
        plt.savefig(f"{eval_dir}/{name}_roc_curve.png", dpi=300)
        plt.close()

        print(f"[INFO] ROC curve saved at {eval_dir}/{name}_roc_curve.png")
    except Exception as e:
        print(f"[WARNING] ROC curve not generated: {e}")
        roc_auc = None

    # --- Save Metrics Summary ---
    metrics_summary = {
        "AUC": roc_auc,
        "Accuracy": report["accuracy"],
        "Precision": report["weighted avg"]["precision"],
        "Recall": report["weighted avg"]["recall"],
        "F1-Score": report["weighted avg"]["f1-score"]
    }

    metrics_path = f"{eval_dir}/{name}_metrics.csv"
    pd.DataFrame([metrics_summary]).to_csv(metrics_path, index=False)
    print(f"[INFO] Metrics summary saved at {metrics_path}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Ensure output folder exists
os.makedirs("results/plots", exist_ok=True)

def plot_training_curves(history, model_name="model"):
    """
    Generate accuracy and loss plots similar to the reference image.
    history: Dictionary containing 'train_acc', 'val_acc', 'train_loss', 'val_loss'
    """

    epochs = np.arange(1, len(history['train_acc']) + 1)

    # ============ Accuracy Plot ============
    plt.figure(figsize=(8, 6))
    plt.plot(epochs, history['train_acc'], 'bo-', label="Training acc")  # 'bo-' = blue dots with line
    plt.plot(epochs, history['val_acc'], 'r-', label="Validation acc")   # 'r-' = red line
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title("(a) Variation of accuracy with epochs")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(f"results/plots/{model_name}_accuracy.png", dpi=300)
    plt.close()

    # ============ Loss Plot ============
    plt.figure(figsize=(8, 6))
    plt.plot(epochs, history['train_loss'], 'bo-', label="Training loss")
    plt.plot(epochs, history['val_loss'], 'r-', label="Validation loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("(b) Categorical cross entropy loss vs number of epochs")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(f"results/plots/{model_name}_loss.png", dpi=300)
    plt.close()

    print(f"âœ… Plots saved in results/plots/ as {model_name}_accuracy.png and {model_name}_loss.png")
```
